```python
import os
import urllib.request

data_location = "Insurance.csv"
data_url = 'https://raw.githubusercontent.com/elephantscale/datasets/master/medical-insurance-cost/insurance.csv'
#data_url = 'https://raw.githubusercontent.com/elephantscale/datasets/master/house-prices/house-sales-full.csv'

if not os.path.exists (data_location):
    data_location = os.path.basename(data_location)
    if not os.path.exists(data_location):
        print("Downloading : ", data_url)
        urllib.request.urlretrieve(data_url, data_location)
print('data_location:', data_location)
```

```python
import pandas as pd
```

```python
pd.options.display.float_format = '{:,.2f}'.format
insurance = pd.read_csv(data_location)
insurance.sample(10)
```

```python
insurance.describe()
```

```python
insurance.info()
```

```python
insurance.corr()
```

```python
insurance['region'].value_counts()
```

```python
ins_new = pd.get_dummies(insurance)
```

```python
ins_new.describe()
```

```python
ins_new.corr()
```

```python
input_columns= ['age','bmi','smoker_no','smoker_yes']
label_column = ['charges']
```

```python
insurance_final = ins_new[input_columns+label_column]
```

```python
insurance_final.sample(10)
```

```python
x = insurance_final[input_columns]
y = insurance_final[label_column]
x['age']=x['age']/x['age'].max()
x['bmi']=x['bmi']/x['bmi'].max()
```

```python
x['bmi'].max()
```

```python
x.sample(10)
```

```python
from sklearn.linear_model import LinearRegression

##  algo + data --> model
algo = LinearRegression ()
model = algo.fit(x, y)  ## TODO : fit (x,y)
model
```

```python
predictions = model.predict(x)
predictions
```

```python
import numpy as np
np.set_printoptions(precision=2, suppress=True)

print ("coefficients: " , model.coef_)
print ("intercept : ", model.intercept_)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
from math import sqrt

## TODO : find R2 by comparing 'y' and 'predictions'
print ("R2 : " , r2_score(y, predictions))

print ("MSE : ", mean_squared_error(y, predictions))
print ("RMSE : ", sqrt(mean_squared_error(y, predictions)) )
```
