```python
import spacy
print ("spacy version ", spacy.__version__)

nlp = spacy.load("en_core_web_lg")
# nlp = spacy.load("en_core_web_md")
```

```python
text="""Apple unveiled the first iPhone in 2007 in San Francisco, California.  
Steve Jobs did a flawless live demonstration of the phone."""

text="A group of 10 employees who are top level executives from Amazon Inc are trying to fix the production issue when the other employees are on leave "

doc = nlp(text)
```

```python
from spacy import displacy

displacy.render(doc, style="ent", jupyter=True)
```
