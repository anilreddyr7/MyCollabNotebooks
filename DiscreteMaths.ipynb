```python
def permutations(lis):
    if len(lis) == 1:
        return [lis]
    output = []
    *front, last = lis
    for perm in permutations(front):
        for i in range(len(perm) + 1):
            new = perm[:i] + [last] + perm[i:]
            output.append(new)
    return sorted(output)
```

```python
permutations([1,2,3])
```

```python
def get_size(list):
  print(f'computing length of {list}')
  if not list:
    return 0
  return 1 + get_size(list[1:])
```

```python
get_size([1,2,3,4,5])
```

```python
def factorila(n):
  assert n>0
  result = 1
  for i in range(1,n):
    result *= i
  return result
```

```python
factorila(30)
```

```python
def isPrime(num):
  assert num > 0
  if num == 1:
    return False
  for d in range(2,num):
    if num%d == 0:
      return False
  return True
  
```

```python
for i in range(1,100):
  if not isPrime(i**2 + i + 41):
    print(i)
```
