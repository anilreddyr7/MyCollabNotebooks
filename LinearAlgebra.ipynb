```python
def zeros_matrix(rows,cols):
  result = []
  while len(result) < rows:
    result.append([])
    while len(result[-1]) < cols:
      result[-1].append(0)
  return result
```

```python
def find_mean(X):
  mean_val=0
  if len(X)==0:
    return mean_val
  for i in X:
    mean_val = mean_val + i
  return mean_val/(len(X))
```

```python
def find_std(X):
  std_value = 0
  variance_value = 0
  for i in X:
    variance_value = variance_value + (i - find_mean(X))**2
  variance_value = variance_value/len(X)
  std_value = variance_value**(1/2)
  return std_value
```

```python
input = [22,24,26,52,45,37,78,90,110]
find_std(input)
```
